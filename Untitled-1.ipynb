```python
import pandas as pd
import torch
pd.options.display.float_format = '{:,.6f}'.format
pd.options.display.float_format = lambda x: f"{x:,.2f}".replace(",", " ").replace(".", ",")

ds1 = pd.read_csv("data/properties.csv")
ds1.drop(['id', 'subproperty_type', 'region', 'province',  'locality', 'latitude', 'longitude', 'heating_type', 'construction_year', 'fl_swimming_pool', 'fl_open_fire',
           'fl_furnished', 'fl_floodzone', 'fl_double_glazing', 'fl_terrace', 'fl_garden', 'equipped_kitchen', 'cadastral_income', 'state_building', 'primary_energy_consumption_sqm' ], 
           axis=1, inplace=True
           )
ds1 = ds1[ds1['property_type'] == 'HOUSE'].drop(['property_type'], axis=1)
ds1
# ds1.groupby('heating_type')['heating_type'].agg('count')
```

```python
ds1.isnull().sum()
```

```python
ds1.dropna(subset=['total_area_sqm', 'nbr_frontages', 'terrace_sqm', 'garden_sqm'], inplace=True)
ds1 = ds1[ds1['epc'] != 'MISSING']
ds1.shape
```

```python
# Update 'surface_land_sqm' where it is 0 with the sum of 'terrace_sqm' and 'garden_sqm'
ds1.loc[ds1['surface_land_sqm'] == 0, 'surface_land_sqm'] = (
    ds1['terrace_sqm'] + ds1['garden_sqm']
)

ds1
```

```python
ds1[(ds1['surface_land_sqm'] != 0) & ( (ds1['garden_sqm'] == 0) | (ds1['terrace_sqm'] == 0))]
```

```python
ds1.drop(['terrace_sqm', 'garden_sqm'], 
           axis=1, inplace=True
           )
ds1
```

```python
ds1['nbr_frontages'].unique()
ds1[ds1.nbr_frontages > 10]
ds1 = ds1[ds1.nbr_frontages < 10]
ds1
```

```python
ds1.loc[:, 'zip_code'] = ds1['zip_code'].astype(str)
ds1.dtypes
```
